```python
# K-Means Clustering on Cities_r2.csv
# Clustering based on total_graduates

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans

# Load dataset
df = pd.read_csv("./cities_r2.csv")

# Encode categorical columns (if any)
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])

# Select feature for clustering
X = df[["total_graduates"]]

# Apply K-Means (3 clusters is typical; change if needed)
kmeans = KMeans(n_clusters=5, random_state=42)
df["Cluster"] = kmeans.fit_predict(X)

print(df[["total_graduates", "Cluster"]])
```
